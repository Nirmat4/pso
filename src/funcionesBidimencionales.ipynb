```python
# colormaps_demos.py

from matplotlib import cm  
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

def plotFunction(limInf,limSup,pts,nameFunction):
  X=np.linspace(limInf[0],limSup[0],pts)
  Y=np.linspace(limInf[1],limSup[1],pts)
  X,Y=np.meshgrid(X,Y)
  if nameFunction == 'sphere':
    Z=X**2+Y**2  
  elif nameFunction == 'rastrigin':
    (X**2-10*np.cos(2*np.pi*X))+(Y**2-10*np.cos(2*np.pi*Y))+20

  fig=plt.figure()
  ax=fig.add_subplot(111,projection='3d')
  surf=ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.hsv,edgecolor='darkred',linewidth=0.1)

  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('f(x,y)')

  plt.savefig('graficaSphere.jpg')
  plt.show()

limInf=[-2,-1]
limSup=[2,2]
pts=100
nameFunction='sphere'
plotFunction(limInf,limSup,pts,nameFunction)
```

```python
from matplotlib import cm  
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np


#Funcion Ackley
def plotFunction(limInf,limSup,pts,nameFunction):
  X=np.linspace(limInf[0],limSup[0],pts)
  Y=np.linspace(limInf[1],limSup[1],pts)
  X,Y=np.meshgrid(X,Y)
  if nameFunction == 'sphere':
    Z=X**2+Y**2  
  elif nameFunction == 'ackley':
    Z=-20*np.exp(-0.2*np.sqrt(0.5*(X**2+Y**2)))-np.exp(0.5*(np.cos(2*np.pi*X)+np.cos(2*np.pi*Y)))+np.e+20

  fig=plt.figure()
  ax=fig.add_subplot(111,projection='3d')
  surf=ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.hsv,edgecolor='darkred',linewidth=0.1)

  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('f(x,y)')

  plt.savefig('graficaAckley.jpg')
  plt.show()

limInf=[-2,-2]
limSup=[2,2]
pts=100
nameFunction='ackley'
plotFunction(limInf,limSup,pts,nameFunction)
```

```python
from matplotlib import cm  
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Funcin Rosenbrock
def rosenbrock(x,y):
    return (1-x)**2+100*(y-x**2)**2

def plotRosenbrock(limInf,limSup,pts):
    X=np.linspace(limInf[0],limSup[0],pts)
    Y=np.linspace(limInf[1],limSup[1],pts)
    X,Y=np.meshgrid(X,Y)
    Z=rosenbrock(X,Y)

    fig=plt.figure()
    ax=fig.add_subplot(111,projection='3d')
    surf=ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.hsv,edgecolor='darkred',linewidth=0.1)

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x,y)')
    ax.set_title('Funcin de Rosenbrock')

    plt.savefig('graficaRosenbrock.jpg')
    plt.show()

limInf=[-2,-2]
limSup=[2,2]
pts=100
plotRosenbrock(limInf,limSup,pts)
```

```python
from matplotlib import cm  
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Funcin Rastrigin
def rastrigin(x,y,A=10):
    return A*2+(x**2-A*np.cos(2*np.pi*x))+(y**2-A*np.cos(2*np.pi*y))

def plotRastrigin(limInf,limSup,pts):
    X=np.linspace(limInf[0],limSup[0],pts)
    Y=np.linspace(limInf[1],limSup[1],pts)
    X,Y=np.meshgrid(X,Y)
    Z=rastrigin(X,Y)

    fig=plt.figure()
    ax=fig.add_subplot(111,projection='3d')
    surf=ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.hsv,edgecolor='darkred',linewidth=0.1)

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x,y)')
    ax.set_title('Funcin de Rastrigin')

    plt.savefig('graficaRastrigin.jpg')
    plt.show()

limInf=[-5.12,-5.12]
limSup=[5.12,5.12]
pts=100
plotRastrigin(limInf,limSup,pts)
```

```python
from matplotlib import cm  
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Funcin de Griewank
def griewank(x,y):
    return (x**2 / 4000)+(y**2 / 4000)-np.cos(x)*np.cos(y / np.sqrt(2))+1

def plotGriewank(limInf,limSup,pts):
    X=np.linspace(limInf[0],limSup[0],pts)
    Y=np.linspace(limInf[1],limSup[1],pts)
    X,Y=np.meshgrid(X,Y)
    Z=griewank(X,Y)

    fig=plt.figure()
    ax=fig.add_subplot(111,projection='3d')
    surf=ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.hsv,edgecolor='darkred',linewidth=0.1)

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x,y)')
    ax.set_title('Funcin de Griewank')

    plt.savefig('graficaGriewank.jpg')
    plt.show()

limInf=[-600,-600]
limSup=[600,600]
pts=100
plotGriewank(limInf,limSup,pts)
```

```python
def buscaMinimo(x1, x2, x3, f):
  if f(x1) >= f(x2) and f(x2) <= f(x3):
    return 1
  else:
    return 0

def exhaustiva(a, b, n, f):
  x1 = a
  deltaX = (b-a)/n
  x2 = x1 + deltaX
  x3 = x2 + deltaX
  iteraciones = 0
  while True:
    iteraciones = iteraciones + 1
    if buscaMinimo(x1,x2,x3, f) == 1:
      print("El mínimo se encuentra entre", x1, " y ", x3 )
      break
    else:
      x1 = x2
      x2 = x3
      x3 = x2 + deltaX
    if x3 > b:
      break
    #else:
      #print("No existe un mínimo en (a,b) o un punto extremo (a ó b) es el mínimo")
  print("Numero de iteraciones", iteraciones)

a = -5
b = 6
a1 = 0.0001
b1 = 5
n = 500
x = np.linspace(a, b, n)

p1 = plot(x, f6(x))
plt.grid()
n = 10000# 1000 -> 904, 100-> 90, 10 -> 9
exhaustiva(a, b, n, f6) #5, 10, 50
```
